````python
import mysql.connector
from mysql.connector import Error
import pandas as pd

def execute_query(query, params=None, many=False):
    try:
        # Connect to MySQL
        connection = mysql.connector.connect(
            host="localhost",
            user="root",
            password="admin",
            database="Northwind_DB"
        )
        
        if connection.is_connected():
            cursor = connection.cursor()
            if many:
                cursor.executemany(query, params)
            else:
                cursor.execute(query, params)
            
            # If the query is a SELECT statement, fetch and return the results as a DataFrame
            if query.strip().upper().startswith("SELECT"):
                results = cursor.fetchall()
                df = pd.DataFrame(results, columns=[desc[0] for desc in cursor.description])
                for row in results:
                    print(row)
                return df
            else:
                # Commit the transaction for INSERT, UPDATE, DELETE queries
                connection.commit()
                return cursor.rowcount
        
    except Error as e:
        print(f"Error: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

# Create the Products table
create_table_query = """
CREATE TABLE IF NOT EXISTS Products (
    ProductID INT,
    ProductName VARCHAR(255),
    SupplierID INT,
    CategoryID INT,
    Unit VARCHAR(255),
    Price DECIMAL(10, 2)
)
"""
execute_query(create_table_query)

# Insert data into the Products table
insert_data_query = """
INSERT INTO Products (ProductID, ProductName, SupplierID, CategoryID, Unit, Price) VALUES (%s, %s, %s, %s, %s, %s)
"""
products_data = [
    (1, 'Chais', 1, 1, '10 boxes x 20 bags', 18.00),
    (2, 'Chang', 1, 1, '24 - 12 oz bottles', 19.00),
    (3, 'Aniseed Syrup', 1, 2, '12 - 550 ml bottles', 10.00),
    (4, "Chef Anton's Cajun Seasoning", 2, 2, '48 - 6 oz jars', 22.00),
    (5, "Chef Anton's Gumbo Mix", 2, 2, '36 boxes', 21.35)
]
execute_query(insert_data_query, products_data, many=True)

print("Products table created and data inserted successfully.")
````

### Views
Definition: A `view` is a virtual table based on the result-set of an SQL statement. It contains rows and columns, just like a real table.
Example: 

```sql
CREATE VIEW CustomerView AS
SELECT CustomerName, City, Country
FROM Customers;
```

````python
# Create the CustomerView view
create_view_query = """
CREATE VIEW CustomerView AS
    SELECT CustomerName, City, Country
FROM Customers;
"""

execute_query(create_view_query)
print("CustomerView created successfully.")
````

````python
# Drop the CustomerView view
drop_view_query = "DROP VIEW IF EXISTS CustomerView;"
execute_query(drop_view_query)
print("CustomerView dropped successfully.")
````

## Functions
Definition: A `function` is a stored program that you can pass parameters into and return a single value. Functions are used to encapsulate reusable logic.

Example

```sql
CREATE FUNCTION GetCustomerCount()
RETURNS INT
BEGIN
    DECLARE customerCount INT;
    SELECT COUNT(*) INTO customerCount FROM Customers;
    RETURN customerCount;
END;

```

````python
# Create the CustomerView view
create_function_query = """
CREATE FUNCTION GetCustomerCount()
RETURNS INT
BEGIN
    DECLARE customerCount INT;
    SELECT COUNT(*) INTO customerCount FROM Customers;
    RETURN customerCount;
END;
"""

execute_query(create_function_query)
print("GetCustomerCount function created successfully.")
````

````python
# Create the GetCustomerCount function
create_function_query = """
CREATE FUNCTION GetCustomerCount()
RETURNS INT
DETERMINISTIC
BEGIN
    DECLARE customerCount INT;
    SELECT COUNT(*) INTO customerCount FROM Customers;
    RETURN customerCount;
END;

"""
execute_query(create_function_query)

print("GetCustomerCount function created successfully.")
````

Set the `log_bin_trust_function_creators variable to 1`: This allows you to create functions without the additional characteristics, but it is less safe. You can set this variable temporarily or permanently.

- **Temporarily** (for the current session):
```sql
SET GLOBAL log_bin_trust_function_creators = 1;
```

- **Permanently** (by editing the MySQL configuration file): Add the following line to your MySQL configuration file (e.g., my.cnf or my.ini):
```sql
[mysqld]
log_bin_trust_function_creators = 1
```
Then Restart mysql server

## Procedures
Definition: A `procedure` is a stored program that you can pass parameters into and can perform actions such as modifying data or returning multiple values.
Example:

```SQL
CREATE PROCEDURE AddCustomer(
    IN p_CustomerName VARCHAR(255),
    IN p_ContactName VARCHAR(255),
    IN p_Address VARCHAR(255),
    IN p_City VARCHAR(255),
    IN p_PostalCode VARCHAR(10),
    IN p_Country VARCHAR(50)
)
BEGIN
    INSERT INTO Customers (CustomerName, ContactName, Address, City, PostalCode, Country)
    VALUES (p_CustomerName, p_ContactName, p_Address, p_City, p_PostalCode, p_Country);
END;
```

````python
# Create the AddCustomer procedure
create_procedure_query = """
CREATE PROCEDURE AddCustomer(
    IN p_CustomerID INT,
    IN p_CustomerName VARCHAR(255),
    IN p_ContactName VARCHAR(255),
    IN p_Address VARCHAR(255),
    IN p_City VARCHAR(255),
    IN p_PostalCode VARCHAR(10),
    IN p_Country VARCHAR(50)
)
BEGIN
    INSERT INTO Customers (CustomerID, CustomerName, ContactName, Address, City, PostalCode, Country)
    VALUES (p_CustomerID, p_CustomerName, p_ContactName, p_Address, p_City, p_PostalCode, p_Country);
END;
"""
execute_query(create_procedure_query)

print("AddCustomer procedure created successfully.")
````

```Materialized View```
Definition: A `materialized view` is a database object that contains the results of a query. Unlike a regular view, which is a virtual table, a materialized view stores the query result physically. This can improve performance for complex queries as the data is precomputed and stored.

Example:

```SQL
CREATE MATERIALIZED VIEW ProductSummary AS
SELECT CategoryID, COUNT(ProductID) AS TotalProducts, AVG(Price) AS AveragePrice
FROM Products
GROUP BY CategoryID;
```

````python
query = """
CREATE MATERIALIZED VIEW ProductSummary AS
SELECT CategoryID, COUNT(ProductID) AS TotalProducts, AVG(Price) AS AveragePrice
FROM Products
GROUP BY CategoryID;
"""
df_result = execute_query(query)
display(df_result)
````

````python
import mysql.connector
from mysql.connector import Error
import pandas as pd
import time

def execute_query(query, params=None, many=False, retries=3):
    attempt = 0
    while attempt < retries:
        try:
            # Connect to MySQL
            connection = mysql.connector.connect(
                host="localhost",
                user="root",
                password="admin",
                database="Northwind_DB"
            )
            
            if connection.is_connected():
                cursor = connection.cursor()
                if many:
                    cursor.executemany(query, params)
                else:
                    cursor.execute(query, params)
                
                # If the query is a SELECT statement, fetch and return the results as a DataFrame
                if query.strip().upper().startswith("SELECT"):
                    results = cursor.fetchall()
                    df = pd.DataFrame(results, columns=[desc[0] for desc in cursor.description])
                    return df
                else:
                    # Commit the transaction for INSERT, UPDATE, DELETE queries
                    connection.commit()
                    return cursor.rowcount
        
        except Error as e:
            print(f"Error: {e}")
            if "Lock wait timeout exceeded" in str(e):
                attempt += 1
                print(f"Retrying... ({attempt}/{retries})")
                time.sleep(5)  # Wait for 5 seconds before retrying
            else:
                break
        finally:
            if connection.is_connected():
                cursor.close()
                connection.close()
    return None

# Data to insert
data_to_insert = [
    (1, 101, 10, 15.00, '2024-11-01'),
    (2, 102, 5, 25.00, '2024-11-02'),
    (3, 101, 7, 15.00, '2024-11-03'),
    (4, 103, 3, 30.00, '2024-11-04'),
    (5, 102, 2, 25.00, '2024-11-05'),
    (6, 104, 8, 20.00, '2024-11-06'),
    (7, 105, 12, 22.50, '2024-11-07'),
    (8, 101, 5, 15.00, '2024-11-08'),
    (9, 103, 10, 30.00, '2024-11-09'),
    (10, 102, 7, 25.00, '2024-11-10')
]

# Insert query
insert_query = """
INSERT INTO Sales (SaleID, ProductID, Quantity, Price, SaleDate) 
VALUES (%s, %s, %s, %s, %s);
"""

# Execute the query
rows_inserted = execute_query(insert_query, params=data_to_insert, many=True)
if rows_inserted:
    print(f"{rows_inserted} rows were inserted into the Sales table.")
else:
    print("No rows were inserted into the Sales table.")
````
